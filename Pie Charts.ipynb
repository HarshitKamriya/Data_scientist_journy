```python
import matplotlib.pyplot as plt
plt.style.use("ggplot") # Choose any style you like
```

```python
#Data
labels = ["Scahin","Sehwag","Kohli","Yuvraj"]
runs = [18000,8000,12000,9500]

plt.title("Carrer Runs of Indina Batsmen")
plt.pie(runs,labels = labels)
```

```python
colors = ['#f20b30','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
```

```python
plt.pie(
    runs,
    labels = labels,
    colors = colors,
    wedgeprops = {'edgecolor':'#4313ec','linewidth':2,'linestyle':"--",'fill':True}
)
```

```python
plt.pie(
    runs,
    labels = labels,
    colors = colors,
    wedgeprops = {'edgecolor':'#4313ec','linewidth':2,'linestyle':"--",'fill':False}
)
```

```python
explode = [0.1,0,0,0]
plt.pie(runs,labels = labels,explode = explode, colors = colors)
plt.title("Exploded Pie Example")
```

```python
explode = [0.1,0,0,0]
plt.pie(runs,labels = labels,explode = explode, colors = colors,shadow = True,autopct='%1.1f%%')
plt.title("Exploded Pie Example")
```

```python
explode = [0.1,0,0,0]
plt.pie(runs,labels = labels,explode = explode, colors = colors,shadow = True, startangle = 100)
plt.title("Exploded Pie Example")
```

```python
# Too many categories
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]


plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
plt.show()
```
