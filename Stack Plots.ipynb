```python
import matplotlib.pyplot as plt
```

```python
activities = ['Studing','Playing','Watching TV','Sleeping']
```

```python
time_spent = [3,2,2,5]
colors = ['blue','red','yellow','pink']
plt.figure(figsize = (6,6))
plt.pie(time_spent,labels = activities,colors = colors,startangle = 90,autopct = '%1.1f%%')
plt.title("Scool going Student")
plt.axis('equal')
```

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['Studying','Playing','Watching TV','Sleeping']
colors = ['skyblue','lightgreen','gold','lightcoral']
plt.figure(figsize = (6,6))
plt.stackplot(days,studying,playing,watching_tv,sleeping,labels = labels,colors = colors, alpha = 0.6)
plt.legend(loc = "upper left")
plt.title("Weekly Activity Tracker")
plt.xlabel("Day")
plt.ylabel("Hours")
```

```python
# Task to plot my weekly schedule
Acitites = ["Coding","Classes","workout","Sleeping"]
days = ["Monday","Tuesday" ,"Wednesday","Thursday","Friday","Saturday","Sunday"]
coding = [9,8,9,10,10,8,9]
classes = [6,7,6,4,5,5,4]
workout = [1,1,2,1,2,1,2]
sleeping = [8,8,7,8,7,8,8] 
colors =  ['skyblue', 'lightgreen', 'gold', 'lightcoral']
plt.figure(figsize = (10,6))
plt.stackplot(days,coding,classes,workout,sleeping,labels = Acitites,colors= colors, alpha = 0.8)
plt.legend(loc = 'upper left')
plt.grid(True)
plt.title("My Weekly Schedule")
plt.xlabel("Days of week")
plt.ylabel("Hours of Day")


```
