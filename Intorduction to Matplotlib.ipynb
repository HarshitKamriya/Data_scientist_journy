```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,6])
plt.show()
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
```

```python
plt.plot(years,runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years,kohli,label = "Virat kohli")
plt.plot(years,sehwag,label = "Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Run Scored")
plt.title("Performance Comparison")
plt.legend()
```

```python
plt.plot(years, kohli, 'ro--',label = "Kohli") # for formatin string color,shape,line style
plt.plot(years,sehwag,'g^:',label = "Sehwag")
plt.legend()
```

```python
# We can also do Color and Line Style Arguments by seperately writing for redability 
# of code
plt.plot(years,kohli,color = 'orange',linestyle = '--' , label = "Kohli")
plt.plot(years,sehwag,color = 'green',linestyle= '-.' , label = "Sehwag")
plt.plot(years,runs,color = 'blue',label = "Tendulkar")
plt.legend()
```

```python
plt.plot(years,kohli,color = 'orange',linestyle = '--' , label = "Kohli")
plt.plot(years,sehwag,color = 'green',linestyle= '-.' , label = "Sehwag")
plt.plot(years,runs,color = 'blue',label = "Tendulkar")
plt.grid(True)
plt.legend()
```

```python
plt.plot(years,kohli,color = 'orange',linestyle = '--' , label = "Kohli")
plt.plot(years,sehwag,color = 'green',linestyle= '-.' , label = "Sehwag")
plt.plot(years,runs,color = 'blue',label = "Tendulkar")
plt.grid(True)
plt.tight_layout()
plt.legend()
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
rohit =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
plt.plot(years,rohit,'co--',label = "Rohit")
plt.plot(years,kohli,'g^-',label = "Kohli")
plt.plot(years,sehwag,'b.:',label = "Sehwag")
plt.xlabel('year')
plt.ylabel('runs')
plt.style.use('dark_background')
plt.grid(True)
plt.legend()

```
