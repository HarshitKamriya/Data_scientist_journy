```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1,2,3,4,5,6,7,8,9]
exam_scores = [40,45,90,55,60,25,75,85,90]
```

```python
colors = ['red' if score<50 else 'green' for score in exam_scores]
sizes = [score*2 for score in exam_scores]
plt.scatter(study_hours,exam_scores,c = colors, s = sizes)
plt.grid(True)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.style.use('ggplot')
plt.title("Study vs Marks")
```

```python
# Sample data
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]

plt.scatter(study_hours, exam_scores)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

```python
# Using a colormap
import numpy as np
scores_normalized = np.array(exam_scores)
plt.scatter(study_hours,exam_scores, c= scores_normalized,cmap = 'viridis')
plt.colorbar(label = "Score")
plt.title("Scatter Plot with Colormap")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.grid(True)
```

```python
# Adding Annotations
plt.scatter(study_hours,exam_scores)
# add labels
for i in range(len(study_hours)):
    plt.annotate(f'S{i+1}',(study_hours[i]+0.2,exam_scores[i]+2.5))

plt.title('Scatter Plot with Annotations')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
```

```python
# Multiple Groups in One Plot
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45,85, 25, 95]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 70, 60, 90]

plt.scatter(class_a_hours,class_a_scores,label = 'A',color = "blue")
plt.scatter(class_b_hours,class_b_scores,label = 'B',color = "red")
plt.title("Scatter Plot: Class A vs Class B")
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
```
